```python
import re

def clean_text(text):
  #remove RT
  text=re.sub(r'RT','',text)
 #Fix &
  text=re.sub(r'&amp:','',text)
  #remove punctuations
  text=re.sub(r'[?!.;:,#@-]','',text)

  text=text.lower()
  return text
```
